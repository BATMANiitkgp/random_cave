```python
#this code computes the correlation coefficient of two random sets
#the function correlation_coeff() returns the value of correlation coefficient (rho) upon execution
#rho = cov(set1,set2)/sqrt(var(set1)*var(set2))
#set1 and set2 are arrays of 10000 elements whose values are randomly assign. 

import math
import random
import numpy as np
def correlation_coeff():
    
    sum1=0
    sum2=0
    sum3=0
    set3=[]
    set1=np.array(random.sample(range(-2,1000),1000))
    
    for i in set1:
        sum1 = sum1+i
        
    set2=np.array(random.sample(range(-100,1000),1000))
    for j in set2:
        sum2=sum2+j
    
    set3=set1*set2
    for k in set3:
        sum3=sum3+k

#mean1 is the mean of set1
#mean2 is the mean of set2
#mean3 is the mean of set3
    mean1=sum1/len(set1)
    mean2=sum2/len(set2)
    mean3=sum3/len(set3)
    
#covariance of set1 and set2 is given by cov12
#covariance = E[XY]-UxUy
    cov12=mean3-(mean1*mean2)

#computing E[X^2]
    mean112=mean1*mean1
    mean222=mean2*mean2
    set11=set1*set1
    set22=set2*set2
    sum11=0
    sum22=0

    for a in set11:
        sum11=sum11+a
    
    for b in set22:
        sum22=sum22+b
    
    mean11=sum11/len(set11)
    mean22=sum22/len(set22)

#variance= E[x^2]-Ux^2
#var11 is the variance of set1
#var22 is the variance of set2
    var11=mean11-mean112
    var22=mean22-mean222
    
#rho is the correlation coefficient
    rho = cov12/math.sqrt(var11*var22)
    print(rho)
    
#call the function correlation_coeff() to print the value of rho
```
